```python
import os
import pandas as pd

# Đường dẫn đến thư mục chứa các file CSV
input_dir = "/kaggle/input/fixlabelcicids2017"
output_path = "/kaggle/working/full_cicids2017.csv"

# Tạo danh sách các file CSV
file_names = [f"{i}.csv" for i in range(1, 9)]

# Đọc và gộp các file CSV
dataframes = []
for file in file_names:
    file_path = os.path.join(input_dir, file)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        dataframes.append(df)
    else:
        print(f"⚠️ Không tìm thấy file: {file_path}")

# Gộp tất cả các DataFrame
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)
    # Lưu DataFrame đã gộp vào thư mục làm việc
    merged_df.to_csv(output_path, index=False)
    print(f"✅ Đã lưu file gộp tại: {output_path}")
else:
    print("❌ Không có file nào được gộp.")
```

```python
import pandas as pd
import numpy as np
import sys
import shutil
import importlib
import os

sys.path.insert(0, '/kaggle/working/relevagan')
from relevagan import train_RELEVAGAN_CC
import datetime

# Load data
Train = pd.read_csv('/kaggle/working/full_cicids2017.csv_(Preprocessed).csv')

# ✅ Fix: Separate data correctly BEFORE any manipulation
label_cols = [col for col in Train.columns if 'Label' in col]
data_cols = [col for col in Train.columns if col not in label_cols]

print("Data columns count (BEFORE processing):", len(data_cols))
print("Total columns:", len(Train.columns))

# ✅ Convert labels properly
Train['Label'] = Train['Label'].apply(lambda x: 0 if str(x).upper() == 'BENIGN' else 1)

print("Train data shape:", Train.shape)
print("Original Label distribution:")
print(Train['Label'].value_counts())

# ✅ Create balanced dataset with CORRECT logic
# RelevGAN expects: Label 0 = bots/attacks, Label 1 = normal
attacks = Train.loc[Train['Label'] == 1].copy()  # DDoS attacks
attacks['Label'] = 0  # Assign as bots for RelevGAN

normal_size = min(10000, attacks.shape[0] // 10)
normal_data = attacks.sample(n=normal_size).copy()
normal_data['Label'] = 1  # Assign as normal for RelevGAN

# Combine
combined_train = pd.concat([attacks, normal_data]).reset_index(drop=True)

print("After fix Label distribution:")
print(combined_train['Label'].value_counts())

normal = combined_train.loc[combined_train['Label'] == 1].copy()  # Normal
bots = combined_train.loc[combined_train['Label'] == 0].copy()    # Bots

print("Normal shape:", normal.shape)
print("Bots shape:", bots.shape)

# ✅ Use combined data for training - VERIFY DIMENSIONS
train = combined_train.copy()

# ✅ CRITICAL: Verify data_cols dimension matches discriminator expectation
print("Final data_cols count:", len(data_cols))
print("Sample data shape without Label:", train[data_cols].shape)

if train.shape[0] > 0:
    # Prepare for batch training
    batch_size = 256
    remaining = train.shape[0] % batch_size
    if remaining > 0:
        additional = batch_size - remaining
        _additional = train.tail(additional)
        train = pd.concat([train, _additional]).reset_index(drop=True)

    print("Final train shape for GAN:", train.shape)
    print("Data dimensions being passed to GAN:", len(data_cols))

    # ✅ Verify no shape mismatch
    sample_batch = train[data_cols].iloc[:batch_size]
    print("Sample batch shape:", sample_batch.shape)
    print("Expected: (256, 68), Actual: (256, {})".format(sample_batch.shape[1]))

    # GAN training parameters
    required_epochs = 3
    nb_steps = required_epochs * train.shape[0] // batch_size
    log_interval = train.shape[0] // batch_size
    
    arguments = [
        100,  # rand_noise_dim
        nb_steps, batch_size, 
        1, 1, 100,  # k_d, k_g, critic_pre_train_steps
        log_interval, 5e-4, 256,  # log_interval, learning_rate, base_n_count
        "/kaggle/working/", "/kaggle/working/",  # CACHE_PATH, FIGS_PATH
        False, train.shape[0], 0,  # show, test_size, gpu_device
        'Accuracy', datetime.datetime.now().strftime("%Y%m%d_%H%M%S"), 'CIC-2017'
    ]

    print("Starting RelevGAN training with balanced data...")
    
    # ✅ CRITICAL: Pass correct dimensions
    print("Discriminator will be initialized with IMG_SHAPE =", len(data_cols))

    # ✅ Train GAN with correctly shaped data
    best_losses = train_RELEVAGAN_CC(arguments, train, combined_train, data_cols)

    print("Training completed!")
    print("Best losses:", best_losses)
else:
    print("❌ No data available for training")
```
